```python
top_states = df.sort_values(by='totalConfirmed', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_states['state'], top_states['totalConfirmed'])
plt.title("Top 10 States by Total Confirmed Cases")
plt.xlabel("Total Confirmed Cases")
plt.ylabel("States")
plt.show()
```

```python
top_deaths = df.sort_values(by='deaths', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_deaths['state'], top_deaths['deaths'])
plt.title("Top 10 States by COVID Deaths")
plt.xlabel("Deaths")
plt.ylabel("States")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df['state'], df['total_Hospitals'])
plt.xticks(rotation=90)
plt.title("Total Hospitals Per State")
plt.ylabel("Hospitals")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['state'], df['rural_Beds'], label='Rural Beds')
plt.plot(df['state'], df['urban_Beds'], label='Urban Beds')
plt.xticks(rotation=90)
plt.legend()
plt.title("Rural vs Urban Beds Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
